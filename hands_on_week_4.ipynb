```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

# 1. Data Encoding

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df_label_encode = df.copy()
df_one_encode = df.copy()
```

```python
#label encod
df_label_encode['Sex']= df_label_encode['Sex'].map({'male':0, 'female':1, 'Laki-Laki':0, 'Perempuan':1})
df_label_encode['Embarked']= df_label_encode['Embarked'].map({'S':0, 'C':1, 'Q':2, 'Southampton':0, 'Cherbourg':1, 'Queenstown':2})
```

# 3. Normalisasi

```python
df.head()
```

```python
scaler = MinMaxScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```

```python
scaler = StandardScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```
